## Choropleth Maps

If you would like to plot geographic boundaries, such as states and countries, you must load the geographic shape data to show in Altair.
This takes a bit of boilerplate (we're thinking about how to streamline this kind of common construct in future releases) and uses the ``geoshape`` marker.

For example, here are the state boundaries:

```python
# Mengimpor library Altair dengan menggunakan alias 'alt'
import altair as alt
# Mengimpor dataset 'world_110m' dari library vega_datasets
from vega_datasets import data

# Mendefinisikan fitur geospasial (countries) menggunakan dataset topojson dari library vega_datasets
countries = alt.topo_feature(data.world_110m.url, 'countries')

# Membuat chart menggunakan Altair dengan menggunakan data fitur geospasial countries
# Jenis mark yang digunakan adalah geoshape (bentuk geometri)
# Warna isi diatur menjadi 'lightgray' dan warna garis tepi diatur menjadi 'white'
# Proyeksi peta yang digunakan adalah 'equirectangular'
# Lebar chart ditetapkan menjadi 500 piksel dan tinggi chart ditetapkan menjadi 300 piksel
chart = alt.Chart(countries).mark_geoshape(
    fill='lightgray',
    stroke='white'
).project(
    "equirectangular"
).properties(
    width=500,
    height=300
)

# Menampilkan chart
chart
```

You can see what happens if you try other projection types; for example, you might try "mercator", "orthographic", "albers", or "gnomonic".

## Points on Background

If you'd like to plot data points on a map background, the easiest way to do so is to use the layering operators we saw in [04-Compound-charts](04-Compound-charts.ipynb).
For example:

```python
# Mengimpor library Altair dengan menggunakan alias 'alt'
import altair as alt
# Mengimpor dataset 'us_10m' dan 'airports' dari library vega_datasets
from vega_datasets import data

# Mendefinisikan fitur topografi (states) menggunakan dataset topojson dari library vega_datasets
states = alt.topo_feature(data.us_10m.url, feature='states')
# Mendapatkan data bandara dari dataset vega_datasets
airports = data.airports()

# Membuat chart latar belakang menggunakan fitur topografi (states)
# Jenis mark yang digunakan adalah geoshape (bentuk geometri)
# Warna isi diatur menjadi 'lightgray' dan warna garis tepi diatur menjadi 'white'
# Proyeksi peta yang digunakan adalah 'albersUsa'
# Lebar chart ditetapkan menjadi 500 piksel dan tinggi chart ditetapkan menjadi 300 piksel
background = alt.Chart(states).mark_geoshape(
    fill='lightgray',
    stroke='white'
).project('albersUsa').properties(
    width=500,
    height=300
)

# Membuat chart titik menggunakan data bandara
# Jenis mark yang digunakan adalah lingkaran (circle)
# Variabel longitude dienkoding sebagai sumbu bujur (longitude) dengan tipe data kuantitatif (Q)
# Variabel latitude dienkoding sebagai sumbu lintang (latitude) dengan tipe data kuantitatif (Q)
# Ukuran lingkaran ditetapkan menjadi 10 piksel
# Tooltip akan menampilkan informasi nama bandara ketika dihover
points = alt.Chart(airports).mark_circle().encode(
    longitude='longitude:Q',
    latitude='latitude:Q',
    size=alt.value(10),
    tooltip='name'
)

# Menggabungkan chart latar belakang dan chart titik
# Hasilnya adalah peta dengan latar belakang berwarna abu-abu muda dan titik-titik yang mewakili lokasi bandara
background + points
```

Notice that we need only specify the projection and the chart size once

## Colored Choropleths

The most complicated type of chart to make is one where the map regions are colored to reflect underlying data; the reason this is complicated is because it often involves joining two different datasets using a *lookup transform*.

Again, this is a piece of the API that we're hoping to improve in the future.

As an example, here is a chart representing the total population of each state:

```python
# Mengimpor dataset 'population_engineers_hurricanes' dari library vega_datasets
pop = data.population_engineers_hurricanes()

# Menampilkan lima baris pertama dari dataframe 'pop'
pop.head()
```

```python
# Mengimpor library Altair dengan menggunakan alias 'alt'
import altair as alt
# Mengimpor dataset 'us_10m' dari library vega_datasets
from vega_datasets import data

# Mendefinisikan fitur topografi (states) menggunakan dataset topojson dari library vega_datasets
states = alt.topo_feature(data.us_10m.url, 'states')

# Mendefinisikan daftar variabel yang akan digunakan untuk warna (population, engineers, hurricanes)
variable_list = ['population', 'engineers', 'hurricanes']

# Membuat chart menggunakan Altair dengan menggunakan data states
# Jenis mark yang digunakan adalah geoshape (bentuk geometri)
# Warna dienkoding berdasarkan variabel population dengan tipe data kuantitatif (Q)
# Transformasi lookup digunakan untuk menggabungkan data populasi, insinyur, dan angin topan berdasarkan ID negara bagian
# Lebar chart ditetapkan menjadi 500 piksel dan tinggi chart ditetapkan menjadi 300 piksel
# Proyeksi peta yang digunakan adalah 'albersUsa'
chart = alt.Chart(states).mark_geoshape().encode(
    color='population:Q'
).transform_lookup(
    lookup='id',
    from_=alt.LookupData(pop, 'id', list(pop.columns))
).properties(
    width=500,
    height=300
).project(
    type='albersUsa'
)

# Menampilkan chart
chart
```
