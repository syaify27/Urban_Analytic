```python
import plotly.graph_objects as go
import numpy as np

# Contoh data grid 3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Membuat peta 3D dengan Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Menampilkan plot
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Contoh data elevasi (misalnya DEM)
# Ini adalah data buatan, Anda dapat menggantinya dengan data aktual
elevation_data = np.random.rand(50, 50) * 100  # Data elevasi acak

# Membuat grid data untuk peta geografis
x = np.linspace(-180, 180, elevation_data.shape[0])
y = np.linspace(-90, 90, elevation_data.shape[1])
x, y = np.meshgrid(x, y)

# Membuat peta 3D dengan Plotly
fig = go.Figure(data=[go.Surface(z=elevation_data, x=x, y=y, colorscale='Earth')])

# Mengatur tata letak
fig.update_layout(title='Peta Elevasi 3D', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

# Menampilkan plot
fig.show()
```

```python
import geopandas as gpd
import plotly.graph_objects as go
import numpy as np

# URL GeoJSON
geojson_url = 'https://github.com/giswqs/streamlit-geospatial/raw/master/data/us_states.geojson'

# Baca data GeoJSON menggunakan GeoPandas
gdf = gpd.read_file(geojson_url)

# Ekstrak geometri (koordinat) dan atribut ALAND
geometry = gdf.geometry
aland = gdf['ALAND']

# Membuat list dari koordinat dan area
x_coords = []
y_coords = []
z_values = []

for geom, area in zip(geometry, aland):
    if geom.geom_type == 'Polygon':
        x, y = geom.exterior.xy
        x_coords.extend(x)
        y_coords.extend(y)
        z_values.extend([area] * len(x))
    elif geom.geom_type == 'MultiPolygon':
        for poly in geom.geoms:
            x, y = poly.exterior.xy
            x_coords.extend(x)
            y_coords.extend(y)
            z_values.extend([area] * len(x))

# Konversi ke array numpy
x = np.array(x_coords)
y = np.array(y_coords)
z = np.array(z_values)

# Membuat peta 3D dengan Plotly
fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, mode='markers',
                                   marker=dict(size=3, color=z, colorscale='Viridis', opacity=0.8))])

# Mengatur tata letak
fig.update_layout(title='Peta 3D Berdasarkan ALAND', autosize=False,
                  width=1000, height=700,
                  scene=dict(
                      xaxis_title='Longitude',
                      yaxis_title='Latitude',
                      zaxis_title='ALAND'
                  ))

# Menampilkan plot
fig.show()
```

```python
import geopandas as gpd
import plotly.graph_objects as go
import numpy as np

# URL GeoJSON
geojson_url = 'https://github.com/giswqs/streamlit-geospatial/raw/master/data/us_states.geojson'

# Baca data GeoJSON menggunakan GeoPandas
gdf = gpd.read_file(geojson_url)

# Ekstrak geometri (koordinat) dan atribut ALAND
geometry = gdf.geometry
aland = gdf['ALAND']

# Menyimpan koordinat dan area dalam list
x_coords = []
y_coords = []
z_values = []

# Loop melalui setiap geometri
for geom, area in zip(geometry, aland):
    if geom.geom_type == 'Polygon':
        x, y = geom.exterior.xy
        x_coords.append(list(x))
        y_coords.append(list(y))
        z_values.append([area] * len(x))
    elif geom.geom_type == 'MultiPolygon':
        for poly in geom.geoms:
            x, y = poly.exterior.xy
            x_coords.append(list(x))
            y_coords.append(list(y))
            z_values.append([area] * len(x))

# Membuat list plotly polygons
polygons = []
for x, y, z in zip(x_coords, y_coords, z_values):
    polygons.append(go.Surface(
        x=[x],
        y=[y],
        z=[z],
        colorscale='Viridis',
        showscale=False,
        opacity=0.8
    ))

# Membuat layout plot
layout = go.Layout(
    title='Peta 3D Berdasarkan ALAND',
    autosize=False,
    width=1000,
    height=700,
    scene=dict(
        xaxis_title='Longitude',
        yaxis_title='Latitude',
        zaxis_title='ALAND'
    )
)

# Membuat figure dan menambahkan polygons
fig = go.Figure(data=polygons, layout=layout)

# Menampilkan plot
fig.show()
```

```python
import geopandas as gpd
import plotly.graph_objects as go
import numpy as np

# URL GeoJSON
geojson_url = 'https://github.com/giswqs/streamlit-geospatial/raw/master/data/us_states.geojson'

# Baca data GeoJSON menggunakan GeoPandas
gdf = gpd.read_file(geojson_url)

# Ekstrak geometri (koordinat) dan atribut ALAND
geometry = gdf.geometry
aland = gdf['ALAND']

# Menyimpan koordinat dan area dalam list
x_coords = []
y_coords = []
z_values = []

# Loop melalui setiap geometri
for geom, area in zip(geometry, aland):
    if geom.geom_type == 'Polygon':
        x, y = geom.exterior.xy
        x_coords.append(list(x))
        y_coords.append(list(y))
        z_values.append([area] * len(x))
    elif geom.geom_type == 'MultiPolygon':
        for poly in geom.geoms:
            x, y = poly.exterior.xy
            x_coords.append(list(x))
            y_coords.append(list(y))
            z_values.append([area] * len(x))

# Membuat array grid dari koordinat dan nilai z (ALAND)
x_flat = np.concatenate(x_coords)
y_flat = np.concatenate(y_coords)
z_flat = np.concatenate(z_values)

# Konversi koordinat dan nilai z ke grid 2D untuk go.Surface
unique_x = np.unique(x_flat)
unique_y = np.unique(y_flat)

x_grid, y_grid = np.meshgrid(unique_x, unique_y)
z_grid = np.zeros_like(x_grid, dtype=float)

for i in range(len(x_flat)):
    x_idx = np.where(unique_x == x_flat[i])[0][0]
    y_idx = np.where(unique_y == y_flat[i])[0][0]
    z_grid[y_idx, x_idx] = z_flat[i]

# Membuat peta 3D dengan Plotly
fig = go.Figure(data=[go.Surface(
    x=x_grid,
    y=y_grid,
    z=z_grid,
    colorscale='Viridis'
)])

# Mengatur tata letak
fig.update_layout(
    title='Peta 3D Berdasarkan ALAND',
    autosize=False,
    width=1000,
    height=700,
    scene=dict(
        xaxis_title='Longitude',
        yaxis_title='Latitude',
        zaxis_title='ALAND'
    )
)

# Menampilkan plot
fig.show()
```
